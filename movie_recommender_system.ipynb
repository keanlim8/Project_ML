```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#import dataset
movie_title_df = pd.read_csv('Movie_Id_Titles')
movie_title_df
```

```python
movie_rating_df = pd.read_csv('u.data', sep='\t', names=['user_id', 'item_id', 'rating', 'timestamp'])
movie_rating_df
```

```python
movie_rating_df.drop(['timestamp'], axis=1, inplace=True)
movie_rating_df
```

```python
#visualize dataset
movie_df_merged = pd.merge(movie_rating_df, movie_title_df, on='item_id')
movie_df_merged
```

```python
rating_df_mean = movie_df_merged.groupby('title')['rating'].describe()['mean']
rating_df_mean
```

```python
rating_df_count = movie_df_merged.groupby('title')['rating'].describe()['count']
rating_df_count
```

```python
rating_mean_count_df = pd.concat([rating_df_mean, rating_df_count], axis=1)
rating_mean_count_df
```

```python
#perform item-based collaborative filtering
movie_df_merged
```

```python
userid_movietitle_matrix = movie_df_merged.pivot_table(index='user_id', columns='title', values='rating')
userid_movietitle_matrix
```

```python
titanic = userid_movietitle_matrix['Titanic (1997)']
titanic.head(20)
```

```python
starwars = userid_movietitle_matrix['Star Wars (1977)']
starwars.head(20)
```

```python
titanic_corr = pd.DataFrame(userid_movietitle_matrix.corrwith(titanic),columns=['Correlations'])
titanic_corr
```

```python
titanic_corr = titanic_corr.join(rating_mean_count_df['count'])
titanic_corr
```

```python
titanic_corr.dropna(inplace=True)
titanic_corr
```

```python
titanic_corr.sort_values('Correlations', ascending=False)
```

```python
titanic_corr[titanic_corr['count']>80].sort_values('Correlations', ascending=False).head(20)
```

```python
starwars_corr = pd.DataFrame(userid_movietitle_matrix.corrwith(starwars),columns=['Correlations'])
starwars_corr
```

```python
starwars_corr = starwars_corr.join(rating_mean_count_df['count'])
starwars_corr
```

```python
starwars_corr.dropna(inplace=True)
starwars_corr
```

```python
starwars_corr[starwars_corr['count']>80].sort_values('Correlations', ascending=False).head(20)
```

```python
#create item-based collaborative filtering on the entire movie rating dataset instead of 1 movie
userid_movietitle_matrix
```

```python
movie_corr = userid_movietitle_matrix.corr(method='pearson', min_periods=80)
movie_corr
```

```python
my_rating = pd.read_csv('My_Ratings.csv')
my_rating
```

```python
my_rating['Movie Name'][0]
```

```python
similar_movie_list = pd.Series()

for i in range(0,2):
    similar_movie = movie_corr[my_rating['Movie Name'][i]].dropna()
    similar_movie = similar_movie.map(lambda x: x* my_rating['Ratings'][i])
    similar_movie_list = similar_movie_list.append(similar_movie)
    
```

```python
similar_movie_list.sort_values(inplace=True, ascending=False)
similar_movie_list
```
